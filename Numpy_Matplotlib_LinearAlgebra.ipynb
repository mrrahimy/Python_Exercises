```python
def XYGenerataor(sample_size:int, normal_noise):
    rn.seed(4)
    Xs= linspace(0,20, sample_size)
    real_slope= 3
    real_intercept =10
    noise= rn.randn(sample_size)*normal_noise
    Ys= real_slope* Xs + real_intercept + noise #ax+b +e
    return Xs, Ys    
```

```python
x,y = XYGenerataor (100, 5) #x is ordeed
```

```python
# numpy math functions
# ایجاد یک آرایه نمونه
arr = np.array([1, 2, 3, 4, 5])

# ۱. توابع پایه‌ای ریاضی
print("Sum:", np.sum(arr))           # جمع عناصر آرایه
print("Product:", np.prod(arr))      # ضرب عناصر آرایه
print("Absolute:", np.abs(arr))      # قدر مطلق عناصر آرایه
print("Power:", np.power(arr, 2))    # توان عناصر آرایه
print("Square Root:", np.sqrt(arr))  # جذر عناصر آرایه

# ۲. توابع مثلثاتی
angles = np.array([0, np.pi/2, np.pi])
print("Sine:", np.sin(angles))       # سینوس
print("Cosine:", np.cos(angles))     # کسینوس
print("Tangent:", np.tan(angles))    # تانژانت

# ۳. توابع نمایی و لگاریتمی
print("Exponential:", np.exp(arr))   # نمایی
print("Natural Log:", np.log(arr))   # لگاریتم طبیعی
print("Base-10 Log:", np.log10(arr)) # لگاریتم پایه 10

# ۴. توابع گرد کردن
arr_float = np.array([1.23, 4.56, 7.89])
print("Round:", np.round(arr_float)) # گرد کردن به نزدیک‌ترین عدد صحیح
print("Floor:", np.floor(arr_float)) # گرد کردن به سمت پایین
print("Ceil:", np.ceil(arr_float))   # گرد کردن به سمت بالا

# ۵. توابع آماری
print("Mean:", np.mean(arr))         # میانگین
print("Median:", np.median(arr))     # میانه
print("Variance:", np.var(arr))      # واریانس
print("Standard Deviation:", np.std(arr))  # انحراف معیار

# ۶. توابع پیشرفته‌تر
print("GCD of 12 and 18:", np.gcd(12, 18))  # بزرگ‌ترین مقسوم‌علیه مشترک
print("LCM of 12 and 18:", np.lcm(12, 18))  # کوچک‌ترین مضرب مشترک
```

```python
from numpy import linalg as lg
# ایجاد ماتریس‌های نمونه
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
a = np.array([3, 4])
b = np.array([1, 2])
u = np.array([1, 2, 3])
v = np.array([4, 5, 6, 7])

result = A.T  # یا np.transpose(A)

# ۱. ضرب ماتریس‌ها
# application: similarity and distance
print("Matrix Multiplication (A @ B):\n", np.dot(A, B))  # یا A @ B
result = np.dot(a, b)  # یا np.inner(u, v) #result scalar #inner is just used for two vecors

# خارجی- اولین عنصر یو، اولین سطر نتیجه میشود
result = np.outer(u, v) # just for vectors, reslut in this case is 3*4 

# عنصر به عنصر Hadamard Product یا Element-wise Multiplication
# application: apply activation functions
result = A * B  # یا np.multiply(A, B)

# ۲. دترمینان ماتریس
#  دترمینان حجم متوازی‌الاضلاعی را نشان می‌دهد که توسط بردارهای سطر یا ستون ماتریس تشکیل می‌شود
# اگر دترمینان یک ماتریس صفر باشد، ماتریس معکوس‌پذیر نیست
# تجزیه ماتریس‌ها: در روش‌هایی مانند تجزیه LU یا تجزیه QR، دترمینان نقش مهمی ایفا می‌کند.
# application: PCA for dimension reduction
print("Determinant of A:", lg.det(A))

# ۳. معکوس ماتریس
print("Inverse of A:\n", lg.inv(A))

# ۴. حل سیستم معادلات خطی Ax = b
print("Solution to Ax = b:", lg.solve(A, b))

# ۵. مقادیر ویژه و بردارهای ویژه
# بردارهای ویژه متناظر با بزرگ‌ترین مقادیر ویژه، جهت‌های اصلی تغییرات داده‌ها را نشان می‌دهند
eigenvalues, eigenvectors = lg.eig(A)
print("Eigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)

# ۱۲. محاسبه توان ماتریس
print("Matrix Power A^2:\n", lg.matrix_power(A, 2))

# ۸. رتبه ماتریس
# تعداد سطرها یا ستون های مستقل --> کاهش ابعاد
# رتبه ماتریس کواریانس فیچرها PCA
print("Rank of A:", lg.matrix_rank(A))

# تجزیه LU 
# تجزیه ماتریس مربعی به دو ماتریس بالا و پایین مثلثی
# برای حل معادلات خطی

# ۶. تجزیه QR - ماتریس R بالا مثلثی و ماتریس Q orthogonal است , inv (Q) = q.T 
# A= Q.R # it is not essential to be a square matrix لزومی به مربعی بودن نداره
# application: dimensionality reduction
# در ماتریس R مقادیری که قدر مطلق بیشتری دارند ویژگی های مهمتری هستند
Q, R = lg.qr(A)
print("Q matrix (QR Decomposition):\n", Q)
print("R matrix (QR Decomposition):\n", R)

# ۷. تجزیه SVD singular value decomposition
# application: dimensionality reduction, data compression and analysis
# A.A^T و A^T.A رو اشکیل میدیم و مقادیر ویژه رو حساب میکنیم
# از مقادیر ویژه رادیکال میگیریم میشه منفرد و نزولی رو ی قطر سیگما میچینیم
# بردارهای ویژ]ه متناظرشون u,v رو شکل میدن و توی ستون هاشون چیده میشن
# U.S میشه کاهش یافته ما
U, S, VT = lg.svd(A)
print("U matrix (SVD):\n", U)
print("Singular Values (SVD):", S)
print("VT matrix (SVD):\n", VT)

# ۹. نرم‌های ماتریس و بردار
# application: regularization L, L2
print("Frobenius Norm of A:", lg.norm(A, 'fro'))  # نرم فروبنیوس- fهمان فرمول l2 کاربرد ماتریس
print("L1(lasso) Norm of b:", lg.norm(b,1))
print("L2 (ridge) Norm of b:", lg.norm(b))    # نرم L2 بردار
linf_norm = np.linalg.norm(v, np.inf) # نرم L∞ بیش‌ترین مقدار مطلق عناصر بردار است.



# ۱۰. اثر ماتریس (Trace)
# جمع عناصر قطر اصلی مثلا
# np.trace(Wt.W)= l2 norm --> ridge
print("Trace of A:", np.trace(A))

# ۱۱. تجزیه چولسکی (Cholesky Decomposition)
C = np.array([[4, 12, -16], [12, 37, -43], [-16, -43, 98]])  # ماتریس مثبت معین
L = lg.cholesky(C)
print("Cholesky Decomposition of C:\n", L)

A_flattened = A.flatten()  # تبدیل ماتریس به بردار
```
